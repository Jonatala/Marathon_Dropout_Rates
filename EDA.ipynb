```python
# function to load json files
def load_json_data(json_file_path):
    """
    Load JSON data from a file.

    Parameters:
    - json_file_path (str): Path to the JSON file.

    Returns:
    - dict: Loaded JSON data.
    """
    with open(json_file_path, 'r') as json_file:
        marathon_data = json.load(json_file)

    return marathon_data
```

```python
combined_df['GROUPED_CATEGORY'] = combined_df['CATEGORY']

def group_categories(category):
    if 'MV' in category:
        return 'Male Veteran'
    elif 'FV' in category:
        return 'Female Veteran'
    elif 'SW' in category:
        return 'Female'
    elif 'SM' in category or 'All_Male' in category:
        return 'Male'
    else:
        return 'Other'

combined_df['GROUPED_CATEGORY'] = combined_df['GROUPED_CATEGORY'].apply(group_categories)
```
